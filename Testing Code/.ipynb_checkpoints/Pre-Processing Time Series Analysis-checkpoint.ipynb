```python
import pandas as pd
def fetch_train(name):
    filename = "./../Dataset/" + name + ".csv"
    df = pd.read_csv(filename)
    return df
train_1 = fetch_train("train_1").fillna(0)
train_2 = fetch_train("train_2").fillna(0)
train_1.head()
```

```python
from matplotlib import rcParams
import numpy
from matplotlib import pyplot as plt
import matplotlib
rcParams['figure.figsize'] = 18,8
y = train_1.loc[1][1:]
plt.plot(y)
plt.xlabel('Date-Time', fontsize=10)
plt.ylabel('Traffic', fontsize=10)
plt.title('Web Traffic- Original data')
plt.show()
```

```python
#first order and second order differencing
first_order = y.diff()
second_order = first_order.diff()
plt.plot(y)
plt.xlabel('Date-Time', fontsize=10)
plt.ylabel('Traffic', fontsize=10)
plt.title('Web Traffic- Original data')
plt.show()
plt.plot(first_order)
plt.xlabel('Date-Time', fontsize=10)
plt.ylabel('Traffic', fontsize=10)
plt.title('Web Traffic- First Order difference')
plt.show()
plt.plot(second_order)
plt.xlabel('Date-Time', fontsize=10)
plt.ylabel('Traffic', fontsize=10)
plt.title('Web Traffic- Second Order difference')
plt.show()
```

```python
import statsmodels.api as sm
series = y
cycle, trend = sm.tsa.filters.hpfilter(series, 50)
fig, ax = plt.subplots(3,1)
ax[0].plot(series)
ax[0].set_title('Actual')
ax[1].plot(trend)
ax[1].set_title('Trend')
ax[2].plot(cycle)
ax[2].set_title('Cycle')
plt.show()
```

```python
ind = pd.to_datetime(y.index)
arr = []
for i in range(len(y)):
    arr.append(y[i])
arr = pd.DataFrame(arr)
arr.index = ind
decomposition = sm.tsa.seasonal_decompose(arr) 
decomposition.plot();
```

```python
#acf and pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(arr);
plot_pacf(arr);
acf_values= sm.tsa.stattools.acf(y)
pacf_values= sm.tsa.stattools.pacf(y)
```

```python
#Distributing data based on language
lang_dict = dict()
for i in range(len(train_1)):
    lang = train_1['Page'][i][train_1['Page'][i].find(".wikipedia")-2:train_1['Page'][i].find(".wikipedia")]
    temp=train_1.loc[i]
    if(lang not in lang_dict.keys()):
        lang_dict[lang] = [temp]
    else:
        lang_dict[lang].append(temp)
```

```python
for i in range(len(train_2)):
    lang = train_2['Page'][i][train_2['Page'][i].find(".wikipedia")-2:train_2['Page'][i].find(".wikipedia")]
    temp=train_2.loc[i]
    if(lang not in lang_dict.keys()):
        lang_dict[lang] = [temp]
    else:
        lang_dict[lang].append(temp)
        
data_lang = []
```

```python
import numpy as np
from matplotlib import pyplot as plt
i=0
for lang in lang_dict.keys():
        data_lang.append([lang])
        for j in range(len(lang_dict[lang])):
            data_lang[i].append(sum(lang_dict[lang][j][1:]))
        i=i+1
stats_lang = []
for lang in data_lang:
    stats_lang.append([lang[0],sum(lang[1:]),len(lang[1:]),np.mean(lang[1:]),np.std(lang[1:])])
import pandas as pd
stats_lang = pd.DataFrame(stats_lang[:8])

index = np.arange(len(stats_lang))
plt.bar(index,stats_lang[:][3])
plt.xlabel('Language', fontsize=10)
plt.ylabel('Mean of web hits', fontsize=10)
plt.xticks(index, stats_lang[:][0], fontsize=10, rotation=30)
plt.title('Web Traffic mean based on language')
plt.show()
```

```python
#Distibuting Data based on access type and type
type_dict = {"all-agent":list(),"spider":list()}
access_dict = {"access_dict":list(),"desktop":list(),"mobile-web":list()}
for i in range(len(train_1)):
    if("all-access" in train_1["Page"][i][train_1['Page'][i].find(".wikipedia"):]):
        access_dict["access_dict"].append(train_1.loc[i])
    if("desktop" in train_1["Page"][i][train_1['Page'][i].find(".wikipedia"):]):
        access_dict["desktop"].append(train_1.loc[i])
    if("mobile-web" in train_1["Page"][i][train_1['Page'][i].find(".wikipedia"):]):
        access_dict["mobile-web"].append(train_1.loc[i]) 
    if("all-agent" in train_1["Page"][i][train_1['Page'][i].find(".wikipedia"):]):
        type_dict["all-agent"].append(train_1.loc[i])
    if("spider" in train_1["Page"][i][train_1['Page'][i].find(".wikipedia"):]):
        type_dict["spider"].append(train_1.loc[i])
```
